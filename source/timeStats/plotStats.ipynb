```python
files = df['File name']
x = list(range(len(df['File name'])))

bar_plots = [
    go.Bar(x=x, y=df['Edges modified'], name='Edges modified', marker=go.bar.Marker(color='#0343df')),
    go.Bar(x=x, y=df['Execution time'], name='Execution time', marker=go.bar.Marker(color='#e50000')),
]

layout = go.Layout(
    title=go.layout.Title(text="Cluster Editing Statistics / Quyen Linh TA", x=0.5),
    yaxis_title="Values",
    xaxis_tickmode="array",
    xaxis_tickvals=list(range(27)),
    xaxis_ticktext=tuple(df['File name'].values),
    )

fig = go.Figure(data=bar_plots, layout=layout)

fig.show()
```
