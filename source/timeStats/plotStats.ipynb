```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
import numpy as np
import time
import timeit
```

```python
df = pd.read_csv('stats.csv')
```

```python
df.head()
```

```python
df.plot(subplots=True, figsize=(20,20))
```

```python
x = df['File name']
y = df['Edges modified']
z = df['Execution time']
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))

ax1.plot(x, y, lw=1, color="blue")
ax1.set_ylabel(r"Edges modified", fontsize=18, color="blue")
ax2 = ax1.twinx()
ax2.plot(x, z, lw=1, color="red")
ax2.set_ylabel(r"Execution time $(s)$", fontsize=18, color="red")
```
